```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd
import time

# Define the website to scrape and path where the chromedriver is located
base_url = 'https://www.vesselfinder.com/vessels?page='
path = 'C:\\Users\\agivc\\Downloads\\chromedriver.exe'

# Set up Service and driver
service = Service(executable_path=path)
driver = webdriver.Chrome(service=service)

# List to store vessel data
vessels_data = []

# Function to scrape data from a single page
def scrape_vessels_from_page():
    try:
        # Find all rows that contain vessel data
        ships = driver.find_elements(By.CSS_SELECTOR, 'tr')

        for ship in ships:
            vessel_name = ''
            vessel_type = ''
            details = {}

            try:
                # Check for vessel name
                name_element = ship.find_elements(By.CLASS_NAME, 'slna')
                if name_element:
                    vessel_name = name_element[0].text

                # Check for vessel type
                type_element = ship.find_elements(By.CLASS_NAME, 'slty')
                if type_element:
                    vessel_type = type_element[0].text

                # Navigate to the vessel detail page
                link_element = ship.find_elements(By.CSS_SELECTOR, 'a.ship-link')
                if link_element:
                    vessel_link = link_element[0].get_attribute('href')
                    driver.get(vessel_link)

                    # Wait for details to load
                    WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.CLASS_NAME, 'aparams')))

                    # Extract additional details
                    detail_rows = driver.find_elements(By.CSS_SELECTOR, '.aparams tr')
                    for row in detail_rows:
                        key = row.find_element(By.CLASS_NAME, 'n3').text
                        value = row.find_element(By.CLASS_NAME, 'v3').text
                        details[key] = value

                    # Add vessel data to the list
                    vessels_data.append({
                        "Vessel Name": vessel_name,
                        "Vessel Type": vessel_type,
                        "Predicted ETA": details.get('Predicted ETA', '-'),
                        "Distance / Time": details.get('Distance / Time', '-'),
                        "Course / Speed": details.get('Course / Speed', '-'),
                        "Current Draught": details.get('Current draught', '-'),
                        "Navigation Status": details.get('Navigation Status', '-'),
                        "Position Received": details.get('Position received', '-'),
                        "IMO / MMSI": details.get('IMO / MMSI', '-'),
                        "Callsign": details.get('Callsign', '-'),
                        "Flag": details.get('Flag', '-'),
                        "Length / Beam": details.get('Length / Beam', '-'),
                    })

                    # Go back to the vessels page
                    driver.back()
                    time.sleep(2)  # Wait for the page to load
                    # Wait for the ships list to be ready again
                    WebDriverWait(driver, 10).until(EC.presence_of_all_elements_located((By.CSS_SELECTOR, 'tr')))
                    
            except Exception as e:
                print(f"Error processing ship: {e}")

    except Exception as e:
        print(f"Error occurred while scraping vessels: {e}")

# Function to scrape multiple pages
def scrape_multiple_pages(num_pages):
    for page in range(1, num_pages + 1):
        print(f"Scraping page {page}...")
        # Navigate to the current page URL
        driver.get(base_url + str(page))

        # Wait for the page to load
        WebDriverWait(driver, 10).until(EC.presence_of_all_elements_located((By.CSS_SELECTOR, 'tr')))
        scrape_vessels_from_page()

# Specify how many pages to scrape
pages_to_scrape = 5  # Change this value to scrape more or fewer pages
scrape_multiple_pages(pages_to_scrape)

# Print found vessel data
print("Data kapal yang ditemukan:")
for data in vessels_data:
    print(data)

# Close the browser
driver.quit()

# Save results to DataFrame and CSV
df = pd.DataFrame(vessels_data)
df.to_csv("C:/Users/agivc/Downloads/vessel_detail_data_hlm.csv", index=False)
print("Data berhasil disimpan ke vessel_detail_data_hlm.csv")
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd
import time

# Define the website to scrape and path where the chromedriver is located
website = 'https://www.vesselfinder.com/vessels'
path = 'C:\\Users\\agivc\\Downloads\\chromedriver.exe'

# Set the number of pages to scrape
num_pages_to_scrape = 10  # Set this to the desired number of pages

# Set up Service and driver
service = Service(executable_path=path)
driver = webdriver.Chrome(service=service)

# Open Google Chrome with chromedriver
driver.get(website)

# Wait for the page to load
WebDriverWait(driver, 10).until(EC.presence_of_all_elements_located((By.CSS_SELECTOR, 'tr')))

# List to store vessel data
vessels_data = []

try:
    for page in range(num_pages_to_scrape):  # Loop through the specified number of pages
        print(f"Scraping page {page + 1}...")  # Log the current page

        # Find all rows that contain vessel data
        ships = driver.find_elements(By.CSS_SELECTOR, 'tr')

        for i in range(len(ships)):
            vessel_name = ''
            vessel_type = ''
            details = {}

            try:
                # Re-fetch ships each iteration to avoid stale reference
                ships = driver.find_elements(By.CSS_SELECTOR, 'tr')
                ship = ships[i]

                # Check for vessel name
                name_element = ship.find_elements(By.CLASS_NAME, 'slna')
                if name_element:
                    vessel_name = name_element[0].text

                # Check for vessel type
                type_element = ship.find_elements(By.CLASS_NAME, 'slty')
                if type_element:
                    vessel_type = type_element[0].text

                # Navigate to the vessel detail page
                link_element = ship.find_elements(By.CSS_SELECTOR, 'a.ship-link')
                if link_element:
                    vessel_link = link_element[0].get_attribute('href')
                    driver.get(vessel_link)

                    # Wait for details to load
                    WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.CLASS_NAME, 'aparams')))

                    # Extract additional details
                    detail_rows = driver.find_elements(By.CSS_SELECTOR, '.aparams tr')
                    for row in detail_rows:
                        key = row.find_element(By.CLASS_NAME, 'n3').text
                        value = row.find_element(By.CLASS_NAME, 'v3').text
                        details[key] = value

                    # Add vessel data to the list
                    vessels_data.append({
                        "Vessel Name": vessel_name,
                        "Vessel Type": vessel_type,
                        "Predicted ETA": details.get('Predicted ETA', '-'),
                        "Distance / Time": details.get('Distance / Time', '-'),
                        "Course / Speed": details.get('Course / Speed', '-'),
                        "Current Draught": details.get('Current draught', '-'),
                        "Navigation Status": details.get('Navigation Status', '-'),
                        "Position Received": details.get('Position received', '-'),
                        "IMO / MMSI": details.get('IMO / MMSI', '-'),
                        "Callsign": details.get('Callsign', '-'),
                        "Flag": details.get('Flag', '-'),
                        "Length / Beam": details.get('Length / Beam', '-'),
                    })

                    # Go back to the vessels page
                    driver.back()
                    time.sleep(2)  # Wait for the page to load
                    # Wait for the ships list to be ready again
                    WebDriverWait(driver, 10).until(EC.presence_of_all_elements_located((By.CSS_SELECTOR, 'tr')))
                
            except Exception as e:
                print(f"Error processing ship at index {i}: {e}")

        # Click on the "Next" button to go to the next page
        try:
            next_button = driver.find_element(By.CSS_SELECTOR, 'a.pagination-next')
            next_button.click()
            time.sleep(2)  # Wait for the next page to load
            WebDriverWait(driver, 10).until(EC.presence_of_all_elements_located((By.CSS_SELECTOR, 'tr')))
        except Exception as e:
            print("No more pages to load or error clicking next:", e)
            break  # Exit the loop if an error occurs

    # Print found vessel data
    print("Data kapal yang ditemukan:")
    for data in vessels_data:
        print(data)

except Exception as e:
    print(f"Error occurred: {e}")

finally:
    # Close the browser
    driver.quit()

# Save results to DataFrame and CSV
df = pd.DataFrame(vessels_data)
df.to_csv("C:/Users/agivc/Downloads/vessel_detail_data_hlm2.csv", index=False)
print("Data berhasil disimpan ke vessel_detail_data_hlm2.csv")
```
